```python
# https://plotly.com/python-api-reference/
import plotly.graph_objects as go
def rendering3D(pts, line=False):
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=[p[0] for p in pts], 
        y=[p[1] for p in pts], 
        z=[p[2] for p in pts], 
        mode='lines+markers' if line else 'markers',
        name='Vector',
        marker=dict(color='blue', size=3),
        line=dict(color='red', width=2),
        showlegend=False 
    ))
    fig.update_layout(
        scene=dict(
            camera=dict(
                eye=dict(x=1.5, y=-1.5, z=1.5), 
                center=dict(x=0, y=0, z=0),
                up=dict(x=0, y=0, z=1)
            )
        )
    )
    fig.show()

import matplotlib.pyplot as plt
def rendering2D(pts, line=False):
    plt.scatter([p[0] for p in pts], [p[1] for p in pts], c='r', marker='o')
    plt.xlabel('X Label')
    plt.ylabel('Y Label')
    plt.title('2D Point Cloud')
    plt.axis('equal')
    plt.show()
```

```python
import plotly.graph_objects as go

def renderingVecs(pts, line=False):
    fig = go.Figure()

    # ì›ì ì—ì„œ ë²¡í„° ëì ìœ¼ë¡œì˜ í™”ì‚´í‘œë¥¼ ê·¸ë¦¬ê¸°
    for p in pts:
        fig.add_trace(go.Cone(
            x=[0],  # í™”ì‚´í‘œ ì‹œìž‘ x (ì›ì )
            y=[0],  # í™”ì‚´í‘œ ì‹œìž‘ y (ì›ì )
            z=[0],  # í™”ì‚´í‘œ ì‹œìž‘ z (ì›ì )
            u=[p[0]],  # ë²¡í„° x ì„±ë¶„
            v=[p[1]],  # ë²¡í„° y ì„±ë¶„
            w=[p[2]],  # ë²¡í„° z ì„±ë¶„
            colorscale=[[0, 'blue'], [1, 'blue']],  # í™”ì‚´í‘œ ìƒ‰ìƒ
            sizemode="scaled",
            sizeref=0.2  # í™”ì‚´í‘œ í¬ê¸° ì¡°ì • (ìž‘ê²Œ)
        ))

    # ì  ë° ì„  ì¶”ê°€ (ì„ íƒì ìœ¼ë¡œ ì‹œê°í™”)
    if line:
        fig.add_trace(go.Scatter3d(
            x=[p[0] for p in pts],
            y=[p[1] for p in pts],
            z=[p[2] for p in pts],
            mode='lines+markers',
            marker=dict(color='red', size=3),
            line=dict(color='black', width=2),
            showlegend=False
        ))

    fig.update_layout(
        scene=dict(
            camera=dict(
                eye=dict(x=1.5, y=-1.5, z=1.5),
                center=dict(x=0, y=0, z=0),
                up=dict(x=0, y=0, z=1)
            )
        )
    )
    fig.show()

# ë²¡í„° ë°ì´í„° ì •ì˜
vectors = [
    [1, 1, 0],   # ë²¡í„° 1
    [0, 1, 2],   # ë²¡í„° 2
    [-1, -1, 1]  # ë²¡í„° 3
]

# í™”ì‚´í‘œë¥¼ í¬í•¨í•œ 3D ì‹œê°í™” í˜¸ì¶œ
renderingVecs(vectors)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ë²¡í„° ë°ì´í„° ìƒì„± (ì˜ˆì‹œ)
vector = np.array([3, 4])  # ì‹œê°í™”í•  ë²¡í„°
origin = np.zeros(2)  # ì›ì 

# ê·¸ëž˜í”„ ì„¤ì •
fig, ax = plt.subplots(figsize=(10, 10))

# ë²¡í„° ê·¸ë¦¬ê¸°
ax.quiver(origin[0], origin[1],
          vector[0], vector[1],
          angles='xy', scale_units='xy', scale=1,
          color='r', width=0.005)

# ì¶• ë²”ìœ„ ì„¤ì •
max_range = np.array([vector[0], vector[1]]).max() * 1.2
ax.set_xlim([-max_range, max_range])
ax.set_ylim([-max_range, max_range])

# ì¶• ë ˆì´ë¸” ì„¤ì •
ax.set_xlabel('X')
ax.set_ylabel('Y')

# ê·¸ë¦¬ë“œ ì¶”ê°€
ax.grid(True)

# ì›ì  í†µê³¼í•˜ëŠ” ì¶• ì¶”ê°€
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)

# íƒ€ì´í‹€ ì„¤ì •
plt.title('2D Vector Visualization')

# ì¢…íš¡ë¹„ 1:1ë¡œ ì„¤ì •
ax.set_aspect('equal')

# ì¸í„°ëž™í‹°ë¸Œ ëª¨ë“œ í™œì„±í™”
plt.ion()

# ì¤Œ ê¸°ëŠ¥ í™œì„±í™”
plt.connect('scroll_event', lambda event: plt.gcf().canvas.toolbar.zoom())

# ê·¸ëž˜í”„ í‘œì‹œ
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

def plot_3d_vector(vector, color='red'):
    # ë²¡í„° ë°ì´í„° ìƒì„±
    x = [0, vector[0]]
    y = [0, vector[1]]
    z = [0, vector[2]]
    
    # ë²¡í„°ì˜ ê¸¸ì´ ê³„ì‚°
    vector_length = np.sqrt(vector[0]**2 + vector[1]**2 + vector[2]**2)
    
    # ë²¡í„°ì˜ ë°©í–¥ (ì •ê·œí™”ëœ ë²¡í„°)
    direction = vector / vector_length
    
    # í™”ì‚´í‘œ í¬ê¸° ì„¤ì • (ë²¡í„° ê¸¸ì´ì— ë¹„ë¡€)
    arrow_length = vector_length * 0.2
    arrow_radius = vector_length * 0.08
    
    # ì¶• ë²”ìœ„ ì„¤ì •
    max_range = max(abs(np.array(vector))) * 1.5
    
    # ê·¸ëž˜í”„ ìƒì„±
    fig = go.Figure()
    
    # ë²¡í„° ì„  ì¶”ê°€
    fig.add_trace(go.Scatter3d(
        x=x, y=y, z=z,
        line=dict(color=color, width=6),
        marker=dict(size=0),
        name=f'Vector ({vector[0]}, {vector[1]}, {vector[2]})'
    ))
    
    # í™”ì‚´í‘œ ë (ì›ë¿”) ì¶”ê°€
    fig.add_trace(go.Cone(
        x=[vector[0] - direction[0]*arrow_length/2],
        y=[vector[1] - direction[1]*arrow_length/2],
        z=[vector[2] - direction[2]*arrow_length/2],
        u=[direction[0]*arrow_length],
        v=[direction[1]*arrow_length],
        w=[direction[2]*arrow_length],
        sizeref=arrow_radius,
        colorscale=[[0, color], [1, color]],
        showscale=False
    ))
    
    # ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(
        scene = dict(
            xaxis = dict(range=[-max_range, max_range]),
            yaxis = dict(range=[-max_range, max_range]),
            zaxis = dict(range=[-max_range, max_range]),
            aspectmode='cube',
            camera=dict(
                up=dict(x=0, y=0, z=1),
                center=dict(x=0, y=0, z=0),
                eye=dict(x=1.5, y=1.5, z=1.5)
            )
        ),
        title='3D Vector Visualization with Arrows',
        showlegend=True,
        width=800,
        height=800
    )
    
    # ì›ì  ì¶”ê°€
    fig.add_trace(go.Scatter3d(
        x=[0], y=[0], z=[0],
        mode='markers',
        marker=dict(size=6, color='black'),
        name='Origin'
    ))
    
    return fig

# ì—¬ëŸ¬ ë²¡í„°ë¥¼ ë™ì‹œì— í‘œì‹œí•˜ëŠ” í•¨ìˆ˜
def plot_multiple_vectors(vectors):
    fig = go.Figure()
    max_range = max(abs(np.array([v[0] for v in vectors]).flatten())) * 1.5
    
    for vector, color in vectors:
        # ë²¡í„° ë°ì´í„°
        x = [0, vector[0]]
        y = [0, vector[1]]
        z = [0, vector[2]]
        
        # ë²¡í„°ì˜ ê¸¸ì´ì™€ ë°©í–¥ ê³„ì‚°
        vector_length = np.sqrt(vector[0]**2 + vector[1]**2 + vector[2]**2)
        direction = vector / vector_length
        
        # í™”ì‚´í‘œ í¬ê¸° ì„¤ì •
        arrow_length = vector_length * 0.2
        arrow_radius = vector_length * 0.08
        
        # ë²¡í„° ì„  ì¶”ê°€
        fig.add_trace(go.Scatter3d(
            x=x, y=y, z=z,
            line=dict(color=color, width=6),
            marker=dict(size=0),
            name=f'Vector ({vector[0]}, {vector[1]}, {vector[2]})'
        ))
        
        # í™”ì‚´í‘œ ë ì¶”ê°€
        fig.add_trace(go.Cone(
            x=[vector[0] - direction[0]*arrow_length/2],
            y=[vector[1] - direction[1]*arrow_length/2],
            z=[vector[2] - direction[2]*arrow_length/2],
            u=[direction[0]*arrow_length],
            v=[direction[1]*arrow_length],
            w=[direction[2]*arrow_length],
            sizeref=arrow_radius,
            colorscale=[[0, color], [1, color]],
            showscale=False
        ))
    
    # ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(
        scene = dict(
            xaxis = dict(range=[-max_range, max_range]),
            yaxis = dict(range=[-max_range, max_range]),
            zaxis = dict(range=[-max_range, max_range]),
            aspectmode='cube'
        ),
        title='Multiple 3D Vectors Visualization with Arrows',
        width=800,
        height=800
    )
    
    # ì›ì  ì¶”ê°€
    fig.add_trace(go.Scatter3d(
        x=[0], y=[0], z=[0],
        mode='markers',
        marker=dict(size=6, color='black'),
        name='Origin'
    ))
    
    return fig

# # ì˜ˆì‹œ ì‹¤í–‰
# # ë‹¨ì¼ ë²¡í„°
# vector = np.array([2, 3, 4])
# fig = plot_3d_vector(vector)
# fig.show()

# ì—¬ëŸ¬ ë²¡í„°
vectors = [
    (np.array([1, 0, 0]), 'red'),
    (np.array([0, 1, 0]), 'green'),
    (np.array([0, 0, 1]), 'blue'),
]
fig = plot_multiple_vectors(vectors)
fig.show()
```

## Parametric Curve (Explicit Curve)

```python
def linear_interpolation(p0, p1, t):
    x = p0[0]+(p1[0]-p0[0])*t
    y = p0[1]+(p1[1]-p0[1])*t
    z = p0[2]+(p1[2]-p0[2])*t
    return [x,y,z]

p0 = [0,0,0]
p1 = [10,10,50]

mp = linear_interpolation(p0, p1, 0.5)
mp
```

```python
pts = []
for i in range(100):
    t = i * 0.01
    pts.append(linear_interpolation(p0, p1, t))

rendering3D(pts)
```

```python
R = 10

interval = math.pi * 2 / 150

pts = []
for i in range(150):
    t = interval * i
    x = R * math.cos(t)
    y = R * math.sin(t)
    pts.append([x,y,0])

rendering3D(pts)
```

```python
R = 20
turns = 3
interval = math.pi * 2 / 150 * turns

pts = []
for i in range(150):
    t = interval * i
    x = R * math.cos(t)
    y = R * math.sin(t)
    pts.append([x,y,i * 0.5])

rendering3D(pts)
```

```python
R = 10
resolution = 20
interval = math.pi * 2 / resolution
phi_interval = math.pi / resolution  # A new interval for the vertical angle

pts = []
for i in range(resolution):
    t = interval * i  # Angle around the circle
    for j in range(75):
        phi = phi_interval * j - math.pi / 2  # Vertical angle (from -pi/2 to pi/2)
        x = R * math.cos(t) * math.cos(phi)
        y = R * math.sin(t) * math.cos(phi)
        z = R * math.sin(phi)  # z is determined by the vertical angle
        pts.append([x, y, z])

rendering3D(pts)
```
