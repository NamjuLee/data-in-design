### function 2D Surface

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def viz(x, y, z):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.9, edgecolor='none')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)


    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Parametric Surface')


    ax.set_box_aspect([1, 1, 1]) 

    plt.tight_layout()
    plt.show()
```

```python
# Parameters
u0 = -10
u1 = 10
v0 = -10
v1 = 10
u_count = 30
v_count = 30

u = np.linspace(u0, u1, u_count)
v = np.linspace(v0, v1, v_count)
u, v = np.meshgrid(u, v)

x = u 
y = v
z = np.sin(u) * np.cos(v) * 0.1

viz(x,y,z)
```

```python

u0 = 0
u1 = 2 * np.pi
v0 = -np.pi/2
v1 = np.pi/2
u_count = 30
v_count = 30


u = np.linspace(u0, u1, u_count)
v = np.linspace(v0, v1, v_count)
u, v = np.meshgrid(u, v)

x = np.cos(u) * np.cos(v)
y = np.sin(u) * np.cos(v)
z = np.sin(v)


viz(x,y,z)
```

```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


def viz(x, y, z):
    x_np = np.array(x)
    y_np = np.array(y)
    z_np = np.array(z)
    
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    surf = ax.plot_surface(x_np, y_np, z_np, cmap='viridis', alpha=0.9, edgecolor='none')
    
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Parametric Surface')
    
    ax.set_box_aspect([1, 1, 1])
    
    plt.tight_layout()
    plt.show()


u0 = -10
u1 = 10
v0 = -10
v1 = 10
u_count = 30
v_count = 30


du = (u1 - u0) / (u_count - 1)
dv = (v1 - v0) / (v_count - 1)

x = []
y = []
z = []


for i in range(u_count):
    x_row = []
    y_row = []
    z_row = []
    
    u = u0 + i * du
    
    for j in range(v_count):
        v = v0 + j * dv
    
        x_val = u
        y_val = v

        # z_val = math.sin(u) * math.cos(v)
        z_val = v
        
        x_row.append(x_val)
        y_row.append(y_val)
        z_row.append(z_val)
    

    x.append(x_row)
    y.append(y_row)
    z.append(z_row)


viz(x, y, z)
```
